```python
import pandas as pd
from IPython.display import display, HTML # For displaying styled DataFrame in environments like Jupyter

# --- Anchor Point and Calculated Original Time ---
P_anchor = 0.99
N_anchor = 16384
Time_anchor_ms = 100 # milliseconds

# Function to calculate speedup using Amdahl's Law
def amdahls_speedup(P, N):
    if P == 1: 
        return N
    return 1 / ((1 - P) + (P / N))

# Calculate the speedup at the anchor point
speedup_at_anchor = amdahls_speedup(P_anchor, N_anchor)

# Calculate the effective original time (T1) in milliseconds
T_original_ms = Time_anchor_ms * speedup_at_anchor
print(f"Calculated Original Time (T1) on 1 processor: {T_original_ms:.2f} ms (~{T_original_ms/1000:.2f} seconds)\n")

# Define the P values (rows)
P_values = [0.5, 0.75, 0.90, 0.95, 0.99, 1.0]

# Define the N values (columns)
N_values = [1, 2, 8, 128, 512, 16384]

# Helper function to format milliseconds into a human-readable string
def format_time_from_ms(ms_value):
    if ms_value is None:
        return ""
    
    # Use scientific notation for extremely small non-zero values
    if ms_value > 0 and ms_value < 0.0001:
        return f"{ms_value:.1e}ms" # e.g., 1.2e-5ms

    if ms_value < 1:
        if ms_value * 1000 < 1: # less than 1 microsecond
            return f"{ms_value * 1_000_000:.2f}ns"
        return f"{ms_value * 1000:.2f}µs"
    
    if ms_value < 1000: # less than 1 second
        return f"{ms_value:.2f}ms"

    # Now convert to seconds and format for larger units
    total_seconds = ms_value / 1000.0
    
    if total_seconds < 60:
        return f"{total_seconds:.2f}s"
    
    minutes = int(total_seconds // 60)
    remaining_seconds = total_seconds % 60
    
    if minutes < 60:
        return f"{minutes}m {remaining_seconds:.1f}s"
    
    hours = int(minutes // 60)
    minutes %= 60
    
    if hours < 24:
        return f"{hours}h {minutes:02}m {remaining_seconds:.0f}s"
    
    days = int(hours // 24)
    hours %= 24
    
    return f"{days}d {hours:02}h {minutes:02}m {remaining_seconds:.0f}s"


# Create a list to hold the rows of our table (storing raw float ms values)
table_rows_float = []

# Iterate through each P value to create a row
for P in P_values:
    row_data_float = [P] # P value as a float
    
    for N in N_values:
        speedup = amdahls_speedup(P, N)
        time_taken_ms = T_original_ms / speedup
        row_data_float.append(time_taken_ms)
        
    table_rows_float.append(row_data_float)

# Define column headers
column_headers = ['P'] + [f'N={n}' for n in N_values]

# Create a Pandas DataFrame
df = pd.DataFrame(table_rows_float, columns=column_headers)

# --- Styling the DataFrame for Presentation ---

# Function to highlight the minimum value in a row (excluding the P column)
def highlight_min_time(s):
    numerical_values = s.drop('P', errors='ignore') 
    is_min = numerical_values == numerical_values.min()
    styles = ['' for _ in s.index]
    for i, col_name in enumerate(s.index):
        if col_name != 'P': # Don't apply highlight to the 'P' column itself
            if col_name in is_min.index and is_min.loc[col_name]:
                styles[i] = 'background-color: #d4edda' # Light green background for min
    return styles


# Create a single dictionary for all formatting
format_mapping = {}
format_mapping['P'] = "{:.2f}" # Format P column to 2 decimal places
for col in df.columns[1:]: # For all N columns, use the custom time formatter
    format_mapping[col] = format_time_from_ms


# Apply styling
styler = df.style.hide(axis="index") \
    .set_caption(f"Amdahl's Law: Computation Time (Anchored at P={P_anchor:.2f}, N={N_anchor} = {Time_anchor_ms}ms)") \
    .format(format_mapping) \
    .set_table_styles([
        # Header styles
        {'selector': 'th', 'props': [
            ('font-weight', 'bold'),
            ('background-color', '#e9ecef'), 
            ('text-align', 'center'),
            ('padding', '8px')
        ]},
        # Caption style
        {'selector': 'caption', 'props': [
            ('font-size', '1.2em'),
            ('font-weight', 'bold'),
            ('text-align', 'center'),
            ('padding', '10px')
        ]},
        # Hover effect for rows (interactive in HTML)
        {'selector': 'tr:hover', 'props': [('background-color', '#f8f9fa')]}, 
    ]) \
    .set_properties(subset=['P'], **{ # Style for the P column
        'text-align': 'center',
        'font-weight': 'bold',
        'background-color': '#f2f2f2' 
    }) \
    .set_properties(subset=df.columns[1:], **{ # Style for N columns
        'text-align': 'center',
        'padding': '8px'
    }) \
    .apply(highlight_min_time, axis=1) \
    .set_table_styles([ # Add subtle borders (applied last to ensure borders are on top)
        {'selector': 'td, th', 'props': [('border', '1px solid #dee2e6')]}
    ], overwrite=False)


# Display the styled DataFrame
display(styler)

# To save as PDF (requires weasyprint or pdfkit as discussed previously)
# html_string = styler.to_html()
# from weasyprint import HTML, CSS
# HTML(string=html_string).write_pdf("amdahls_table_ms_anchor.pdf")
# print("Table saved as amdahls_table_ms_anchor.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the execution time function for Gustafson's Law (Weak Scaling)
def execution_time_gustafson(S_sequential_fraction, N, T_ref=1.0):
    """
    Calculates execution time according to Gustafson's Law (weak scaling).
    T_ref is the ideal execution time for the scaled problem on N processors.
    In ideal weak scaling, execution time is constant, regardless of N, 
    as long as the sequential fraction (S_sequential_fraction) is constant.
    The value of S_sequential_fraction is the portion of runtime on the N-processor system that is sequential.
    """
    # For Gustafson's Law (weak scaling), the ideal execution time for the scaled problem
    # is a constant, assuming the sequential fraction S is constant on N processors.
    # The different S values define how much "larger" problem you can solve within this time.
    # But the time itself is ideally T_ref.
    return np.full_like(N, T_ref) # Returns an array of T_ref with the same shape as N


# 2. Define the range for N (Number of Processors) for plotting dense curves
N_values_dense = np.linspace(1, 65536, 500)

# 3. Define the N values for specific integer tick labels (powers of two)
power_of_two_exponents = np.arange(0, 17) # 0 to 16, as 2^16 = 65536
N_values_discrete = 2**power_of_two_exponents

# 4. Define the P values (from Amdahl's context), which translate to S for Gustafson
P_values_amdahl = [1.0, 0.90]
# Corresponding S (sequential fraction of runtime on parallel system) for Gustafson's Law
# (If P is Amdahl's parallel fraction, then S_gustafson = 1 - P)
S_values_gustafson = [1 - p for p in P_values_amdahl] # [0.0, 0.02]

# 5. Create the plot
plt.figure(figsize=(14, 8))

# Determine the x-position for the asymptote labels (slightly beyond the right edge of the plot)
max_N_for_plot = 65536
x_label_pos = max_N_for_plot * 1.03 

# Loop through each S value and plot the execution time curve
for i, S_val in enumerate(S_values_gustafson):
    # For ideal Gustafson's Law (weak scaling), execution time is constant (T_ref = 1.0)
    times = execution_time_gustafson(S_val, N_values_discrete, T_ref=1.0)
    
    # Label refers to the Amdahl's P value for consistency with user's input
    label_text = f'P={P_values_amdahl[i]:.2f} (S={S_val:.2f})'
    
    # Plot the function line - it will be a flat line
    line, = plt.plot(N_values_discrete, times, label=label_text, 
                     linestyle='-' if S_val == 0.0 else '--', # Use different line styles
                     linewidth=2)
    
    # Add a text label for the execution time (which is 1.0 for all)
    # We place labels slightly offset vertically to distinguish them, even if values are identical
    vertical_offset = 0.01 if S_val == 0.0 else -0.01 # Small offset for clarity
    
    plt.text(x_label_pos, times[-1] + vertical_offset, 
             f'{times[-1]:.1f}', # Format as relative time
             color=line.get_color(), 
             ha='left', va='center', 
             fontsize=24, 
             fontweight='bold')

# 6. Customize the plot
plt.xscale('log') 

# Set custom x-axis ticks and labels
plt.xticks(N_values_discrete, [str(n) for n in N_values_discrete], fontsize=24)
plt.tick_params(axis='x', rotation=45) 
plt.yticks(fontsize=24)

plt.xlabel('Number of Processors (N)', fontsize=24)
plt.ylabel('Relative Execution Time', fontsize=24)
plt.title("Gustafson's Law (Weak Scaling): Execution Time vs. Number of Processors", fontsize=24)
plt.grid(True, which="both", ls="--", c='0.7') 
plt.legend(title='Parallel Fraction (P) / Sequential Fraction (S)', loc='upper right', fontsize=24)

plt.ylim(0.9, 1.1) # Set y-axis limits to clearly show the flat lines around 1.0
plt.xlim(1, max_N_for_plot * 1.05) # Extend x-axis limit for labels

plt.tight_layout() 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the Scaled Speedup function for Gustafson's Law
def scaled_speedup_gustafson(S_sequential_fraction, N):
    """
    Calculates Scaled Speedup according to Gustafson's Law.
    S_sequential_fraction is the portion of runtime on the N-processor system that is sequential.
    """
    # Ensure N is an array for element-wise operations
    N = np.asarray(N)
    
    # Gustafson's Law: Scaled Speedup = S + N * (1 - S)
    return S_sequential_fraction + N * (1 - S_sequential_fraction)


# 2. Define the range for N (Number of Processors) for plotting dense curves
N_values_dense = np.linspace(1, 65536, 500)

# 3. Define the N values for specific integer tick labels (powers of two)
power_of_two_exponents = np.arange(0, 17) # 0 to 16, as 2^16 = 65536
N_values_discrete = 2**power_of_two_exponents

# 4. Define the P values (from Amdahl's context), and derive S for Gustafson
P_values_amdahl = [0.5, 0.75, 0.90, 0.95, 0.99]
# Corresponding S (sequential fraction of runtime on parallel system) for Gustafson's Law
# S_gustafson = 1 - P_amdahl
S_values_gustafson = [1 - p for p in P_values_amdahl] # [0.5, 0.25, 0.10, 0.05, 0.01]

# 5. Create the plot
plt.figure(figsize=(14, 8))

# Determine the x-position for the speedup labels (slightly beyond the right edge of the plot)
max_N_for_plot = 65536
x_label_pos = max_N_for_plot * 1.01 # Place label 1% beyond the max N

# Loop through each S value (derived from P) and plot the scaled speedup curve
for i, S_val in enumerate(S_values_gustafson):
    scaled_speedups = scaled_speedup_gustafson(S_val, N_values_discrete)
    
    # Label refers to the Amdahl's P value (user's input) and Gustafson's S value
    label_text = f'P={P_values_amdahl[i]:.2f} (S={S_val:.2f})'
    
    # Plot the function line - it will be an increasing, nearly linear line
    line, = plt.plot(N_values_discrete, scaled_speedups, label=label_text, linewidth=2)
    
    # Add a text label for the end speedup value
    plt.text(x_label_pos, scaled_speedups[-1], 
             f'{scaled_speedups[-1]:.0f}x', # Format as integer speedup with 'x' suffix
             color=line.get_color(), 
             ha='left', va='center', 
             fontsize=11, 
             fontweight='bold')


# 6. Customize the plot
plt.xscale('log') # Keep x-axis on a logarithmic scale

# Set custom x-axis ticks and labels
plt.xticks(N_values_discrete, [str(n) for n in N_values_discrete])
plt.tick_params(axis='x', rotation=45) 

plt.xlabel('Number of Processors (N)', fontsize=12)
plt.ylabel('Scaled Speedup (S + N * (1-S))', fontsize=12) # Explicitly stating the formula
plt.title("Gustafson's Law: Scaled Speedup vs. Number of Processors\n(Problem size scales proportionally with N)", fontsize=14)
plt.grid(True, which="both", ls="--", c='0.7') 
plt.legend(title='Parallel Fraction (P) / Sequential Fraction (S)', loc='upper left', fontsize=10)

# Set y-axis limits to show the near-linear growth clearly
# Max speedup will be for S=0.01 (P=0.99) at N=65536 -> 0.01 + 65536 * 0.99 approx 64880
plt.ylim(0, N_values_discrete[-1] * 1.01) # Slightly above max speedup for visibility
#plt.ylim(0,128)
plt.xlim(1, max_N_for_plot * 1.05) # Extend x-axis limit for labels

plt.tight_layout() 
plt.show()
```
